```python
Recommended_list = [[] for _ in range(len(basicinfo))]
row = 0
#pd.DateFrame(,columns = ['price', 'room and size', 'address' , 'website'])
for each in basicinfo:
    for eachinfo in each:
        Recommended_list[row].append(eachinfo.get_text(separator=','))
        #print(each.get_text(separator=' '))
    for a in each.find_all('a', href=True):
        Recommended_list[row].append('redfin.com' + a['href'])
        print('Found the URL:', 'redfin.com' + a['href'])
    row += 1
    print('row is:', row)
```

```python
recomlist = Recommended_list[:,[0, 2, 3, 6]]
recompd = pd.DataFrame(data = recomlist, columns=[['price', 'room and size', 'address' , 'website']])
recomlist
```

```python
recomlistv2 = recomlist.tolist()

for each in recomlistv2:
    print("extracting info for:",each)
    web = 'http://' + each[3]
    response = get(web, timeout = 30, headers=headers)
    soup = BeautifulSoup(response.text, 'lxml')
    basics = soup.find_all('div', {'class': 'keyDetailsList'})
    calculator = soup.find_all('div', {'class': 'CalculatorSummary'})
    goodschool = soup.find_all('tr', {'class': 'schools-table-row'})
    rentalestimate = soup.find_all('div', {'class': 'rentalEstimateStats'})
    
    spec = [[]]
    for i in basics:
        for eachinfo in i:
            #print(eachinfo.get_text(separator=': '),'\n')
            spec[0].append(eachinfo.get_text(separator=': '))
    for i in calculator:
        spec[0].append(i.get_text(separator=': '))
    if len(rentalestimate) > 0:
        spec[0].append(rentalestimate[0].text)
    for i in goodschool:
        #print(i.get_text(separator=' '),'\n')
        spec[0].append(i.text)

    spec = np.array(spec)
    temp = spec[:,15:]
    count = 0
    sumx = 0
    sumy = 0
    for i in temp[0]:
        #print(i[0])
        if i[0].isnumeric():
            x = str(i).split('/', 1)[0]
            sumx += int(x)
            count += 1
    if count > 0:        
        avgschrate = round(sumx/count,2)
    each.append(avgschrate)

    #print(spec[0])
    if 'Rental estimate' in str(spec[0]):
        temp2 = [match for match in spec[0] if 'Rental estimate' in match]
        rentalestimate = str(temp2[0]).split('/', 1)[0]
        renttem = str(rentalestimate).split('-')
        for j in renttem:
            y = j.replace('$', '').replace(',','').replace(' ','')
            sumy += int(y)
            avgrent = sumy/2
        #print(rentalestimate, avgrent)
        each.append(avgrent)
    else:
        each.append('N/A')
        
    if 'Find a Lender' in str(spec[0]):
        temp3 = [match for match in spec[0] if 'Find a Lender' in match]
        caltemp = str(temp3[0]).split(':', 10)
        #print(len(caltemp))
        if len(caltemp) > 8:
            roughmp = caltemp[0]
            #print(caltemp)
            proptax = caltemp[5]
            hoa = caltemp[7]
    unitprice = [match for match in spec[0] if 'Price/Sq.Ft' in match]
    status = [match for match in spec[0] if 'Status' in match]
    daysonredfin = [match for match in spec[0] if 'Time on Redfin' in match]
    yearbuilt = [match for match in spec[0] if 'Year Built' in match]
    lotsize = [match for match in spec[0] if 'Lot Size' in match]
    
    each.append(unitprice[0].split(':',2)[1]) 
    each.append(status[0].split(':',2)[1])
    each.append(daysonredfin[0].split(':',2)[1])
    each.append(yearbuilt[0].split(':',2)[1])
    if lotsize:
        each.append(lotsize[0].split(':',2)[1])
    else:
        each.append('N/A')
    each.append(hoa)
    each.append(proptax)
    each.append(roughmp)
    #print(caltemp)
```
